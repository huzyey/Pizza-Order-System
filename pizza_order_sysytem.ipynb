```python
# Listeler
available_pizzas = ['Margarita', 'Sucuk-Mısır', 'BBQ Soslu Tavuk', 'Karışık', 'Vejeteryan']
pizza_prices = {'Margarita': 5, 'Sucuk-Mısır': 7, 'BBQ Soslu Tavuk': 6, 'Karışık': 8, 'Vejeteryan': 6.5}
available_toppings = ['Mantar', 'Soğan', 'Yeşil Biber', 'Ekstra Peynir']
topping_prices = {'Mantar':1, 'Soğan': 2, 'Yeşil Biber':3, 'Ekstra Peynir':4}
sub_total = []
final_order = {}
customer_adress = {}


# Sipariş
print("Merhaba, pizza sipariş sistemimize hoşgeldiniz! \n")
order_pizza = True
while order_pizza:    
    print("Pizzalarımız: ")
    print()
    for pizzas in available_pizzas:
        print(pizzas)
        print()
    while True:
        pizza = input("Lütfen istediğiniz pizzayı yazınız. \n \n")
        if pizza in available_pizzas:
            print(f"\n Seçtiğiniz pizza: {pizza}. \n")
            sub_total.append(pizza_prices[pizza])
            break
        if pizza not in available_pizzas:
            print(f"Üzgünüz, {pizza} pizza sistemimizde bulunmamaktadır. \n \n")

# Ekstra Malzeme
    order_topping = True
    print("Ekstra malzemeler: \n")
    for toppings in available_toppings:
        print(toppings)
        print()
    while order_topping:
        extra_topping = input("\n Ekstra malzeme ister misiniz? (Evet veya Hayır) \n")
        if extra_topping == "Evet":
            topping = input("\n İstediğiniz ekstra malzemeyi yazınız: \n")
            if topping in available_toppings:
                final_order.setdefault(pizza, [])
                final_order[pizza].append(topping)
                print(f"\n Pizzanıza {topping} eklenmiştir.")
                sub_total.append(topping_prices[topping])
            else:
                print(f"\n Üzgünüz, {topping} sistemimizde bulunmamaktadır. \n")
        elif extra_topping == "Hayır":
            break

        for key, value in final_order.items():
            print(f"\n Siparişiniz: {key}. \n Ekstra malzemeler: {value}. \n")
        check_order = True
        while check_order:
            print()
            order_correct = input("Siparişleriniz Doğru mu? (Evet veya Hayır) \n")
            if order_correct == "Evet":
                check_order = False
                order_pizza = False
            if order_correct == "Hayır":
                print(final_order)
                add_remove = input("\n Eklemek veya Çıkarmak istediğiniz bir ürün var mı? (Ekle veya Çıkar) \n")
                if add_remove == "Çıkar":
                    remove = input("\n Hangi ürünü çıkarmak istersiniz? \n")
                    del final_order[remove]
                    print(final_order)
                if add_remove == "Ekle":
                    check_order = False

# Sipariş
print(f"\n Toplam sipariş tutarınız: TL {sum(sub_total)}")

print("\n Lütfen isim, telefon numarası,adres ve kredi kartı numaranızı yazınız. \n")
customer_adress['name'] = input("İsminiz:")
customer_adress['street_name'] = input("Adresiniz:")
customer_adress['card'] = input("Kredi kartı numarası:")
customer_adress['phonenumber'] = input("Telefon numaranız:")
print()
print(f"Siparişiniz için teşekkür ederiz {customer_adress['name']}.")
print()
print("Siparişinizi en kısa zamanda teslim edeceğiz.")
print()
print(customer_adress['street_name'])
print()
print(f"Herhangi bir sorun durumunda {customer_adress['phonenumber']} numaralı telefondan sizinle iletişime geçeceğiz.")
```
